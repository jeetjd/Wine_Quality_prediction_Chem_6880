```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import sklearn

data=pd.read_csv(r'C:\Users\Max\Documents\red_wine.csv', header=None)

data_np=data.to_numpy()

np.delete(data_np,0,0)

data_np.shape
```

```python
#AdaBoostClassifier
import pandas as pd
import numpy as np
import sklearn
from sklearn.ensemble import AdaBoostClassifier

X=data_np[1:1399,0:11] #put first 4 columns into x
y=data_np[1:1399,-1] #put last column into y

#to change test input, change first value in brackets to the (row before):(row desired)
y_pred=data_np[1399:1599,0:11] #prediction values

ada_class=AdaBoostClassifier() #setup adaboost classifier
ada_class.fit(X,y) #train model

ada_mse=0
j=0
k=1

for i in y_pred:#for size of y_pred
    j=k-1 #counter
    x=ada_class.predict(y_pred[j:k,0:11]) #prediction
    x=x.astype(np.int64) #create an integer
    x=x.item(0) #take out of array
    l=1399+k #counter
    z=data_np[l,11] #corresponding z value
    z=int(z) #convert to integer
    w=x-z #subtract values
    w=w*w #square values; using ^2 doesnt work as well as just multiplying
    ada_mse=ada_mse+w #add to total value
    k=k+1 #increment
ada_mse=ada_mse/200 #divide by N predictions for MSE value
print(ada_mse) #predicted value
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbm_class=GradientBoostingClassifier() #setup classifier
gbm_class.fit(X,y) #train model

gbm_mse=0
j=0
k=1

for i in y_pred:
    j=k-1
    x=gbm_class.predict(y_pred[j:k,0:11])
    x=x.astype(np.int64)
    x=x.item(0)
    l=1399+k
    z=data_np[l,11]
    z=int(z)
    w=x-z
    w=w*w
    gbm_mse=gbm_mse+w
    k=k+1
gbm_mse=gbm_mse/200
print(gbm_mse) #prediction
```

```python
import xgboost as xgb
xtrain=xgb.DMatrix(X,label=y) #send to DMatrix
param={} #setup default parameters, these parameters provide lowest MSE

bst=xgb.train(param,xtrain) #train model

#dtest=xgb.DMatrix(y_pred) #test model
#xg_pred=bst.predict(dtest)
#xg_pred=np.round(xg_pred)

xg_mse=0
j=0
k=1

for i in y_pred:
    j=k-1
    dtest=xgb.DMatrix(y_pred[j:k,0:11]) #test model
    xg_pred=bst.predict(dtest) #predict
    xg_pred=np.round(xg_pred) #have to round value since returns decimal
    x=xg_pred.astype(np.int64)
    x=x.item(0)
    l=1399+k
    z=data_np[l,11]
    z=int(z)
    w=x-z
    w=w*w
    xg_mse=xg_mse+w
    k=k+1
xg_mse=xg_mse/200
print(xg_mse) #prediction
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_class=RandomForestClassifier() #classify random forest
rf_class.fit(X,y) #train model

rf_mse=0
j=0
k=1

for i in y_pred:
    j=k-1
    x=rf_class.predict(y_pred[j:k,0:11])
    x=x.astype(np.int64)
    x=x.item(0)
    l=1399+k
    z=data_np[l,11]
    z=int(z)
    w=x-z
    w=w*w
    rf_mse=rf_mse+w
    k=k+1
rf_mse=rf_mse/200
print(rf_mse)
```
