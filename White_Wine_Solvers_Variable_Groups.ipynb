```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import sklearn

data=pd.read_csv(r'white_wine.csv', header=None)

data_np=data.to_numpy()

np.delete(data_np,0,0)

data_np.shape
```

```python
from sklearn.svm import SVR

#Acidity
X=data_np[1:4798,[1,2,3,9]]
X=np.append(X,data_np[1:4798,[1,2,3,9]],axis=1)
y=data_np[1:4798,-1] #put last column into y

#to change test input, change first value in brackets to the (row before):(row desired)
y_pred=data_np[4798:4898,[1,2,3,9]] #prediction x value
y_pred=np.append(y_pred,data_np[4798:4898,[1,2,3,9]],axis=1)

svr_data=SVR(kernel="rbf",epsilon=.001) #setup SVR with rbf kernel
svr_data.fit(X,y) #train SVR model

#svr_pred=svr_data.predict(y_pred)
#svr_pred=np.round(svr_pred)

svr_mse=0
j=0
k=1

for i in y_pred:
    j=k-1
    svr_pred=svr_data.predict(y_pred[j:k])
    x=np.round(svr_pred)
    x=x.astype(np.int64)
    x=x.item(0)
    l=4798+k
    z=data_np[l,11]
    z=int(z)
    w=x-z
    w=w*w
    svr_mse=svr_mse+w
    k=k+1
svr_mse=svr_mse/100
print(svr_mse) #predicted value
```
